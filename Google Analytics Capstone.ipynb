# Google Data Analytics Capstone Project

### A). Intoduction:

This case study is the Capstone Project of Google Data Analytics Professional Certificate . In this case study I am working as a junior data analyst in the marketing analyst team at Cyclistic, a fictional bike-share company in Chicago.

Cyclistic is a bike-share program that features more than 5,800 bicycles and 600 docking stations. Cyclistic sets itself apart by also offering reclining bikes, hand tricycles, and cargo bikes, making bike-share more inclusive to people with disabilities and riders who can’t use a standard two-wheeled bike. The majority of riders opt for traditional bikes; about 8% of riders use the assistive options. Cyclistic users are more likely to ride for leisure, but about 30% use them to commute to work each day.

In 2016, Cyclistic launched a successful bike-share offering. Since then, the program has grown to a fleet of 5,824 bicycles that are geotracked and locked into a network of 692 stations across Chicago. The bikes can be unlocked from one station and returned to any other station in the system anytime.


### B). Problem Statement:

The director of marketing believes the company’s future success depends on maximizing the number of annual memberships. Therefore, my team wants to understand how casual riders and annual members use Cyclistic bikes differently. From the insights, our team will design a new marketing strategy to convert casual riders into annual members. But first, Cyclistic executives must approve our recommendations, so they must be backed up with compelling data insights and professional data visualizations.

There are 3 pricing plans: single-ride passes, full-day passes, and annual memberships. Customers who purchase single-ride or full-day passes are referred to as Casual riders. Customers who purchase annual memberships are Cyclistic members.

```python
# Importing all the packages

import pandas as pd
import os
import seaborn as sns
from pandas.api.types import CategoricalDtype
from matplotlib import pyplot as plt
```

#### Data Preprocessing

```python
# Importing all the data files using pandas

df_2020_Oct = pd.read_csv('2020_10-divvy-tripdata.csv')
df_2020_Nov = pd.read_csv('2020_11-divvy-tripdata.csv')
df_2020_Dec = pd.read_csv('2020_12-divvy-tripdata.csv')
df_2021_Jan = pd.read_csv('2021_01-divvy-tripdata.csv')
df_2021_Feb = pd.read_csv('2021_02-divvy-tripdata.csv')
df_2021_Mar = pd.read_csv('2021_03-divvy-tripdata.csv')
df_2021_Apr = pd.read_csv('2021_04-divvy-tripdata.csv')
df_2021_May = pd.read_csv('2021_05-divvy-tripdata.csv')
df_2021_June = pd.read_csv('2021_06-divvy-tripdata.csv')
df_2021_July = pd.read_csv('2021_07-divvy-tripdata.csv')
df_2021_Aug = pd.read_csv('2021_08-divvy-tripdata.csv')
df_2021_Sept = pd.read_csv('2021_09-divvy-tripdata.csv')

# Importing all csv files
```

```python
# Merging all the files under variable name df1

df1 = pd.concat([df_2020_Oct,df_2020_Nov,df_2020_Dec,df_2021_Jan,df_2021_Feb,df_2021_Mar,df_2021_Apr,df_2021_May,df_2021_June,df_2021_July,df_2021_Aug,df_2021_Sept], ignore_index = True)
df1
```

```python
# See the Statistical part of our dataset
df1.describe()
```

```python
# Checking out the null values if any

df1.isnull().sum()
```

```python
# Checking out any duplicate values in our dataset
df1.duplicated().any()
```

```python
# Dropping of columns which are not necessary for our analysis

df1 = df1.drop(columns=["start_station_name","start_station_id","end_station_name","end_station_id","end_lat","end_lng"])
df1.tail()
```

```python
# To know the datatype 
df1.info()
```

```python
# Changing the datatype

df1['started_at'] = df1['started_at'].astype('datetime64')
df1['ended_at'] = df1['ended_at'].astype('datetime64')
```

```python
# Rounding off and changing the datatype

df1['ride_length'] = (df1['ended_at'] - df1['started_at'])/pd.Timedelta(minutes=1)
df1['ride_length'] = df1['ride_length'].astype('int32')
```

```python
df1.head()
```

```python
df1.sort_values(by = 'ride_length')
```

```python
# This shows that majority of customers prefer very less ride distant time

df1[df1['ride_length'] > 2].count()     # 48,65,409 people drove the bike less then 2 mins i.e., - 94.73%
```

```python
df1 = df1[df1['ride_length'] >= 2]
df1 = df1.reset_index()
df1 = df1.drop(columns=['index'])

df1
```

```python
# Visualizing using seaborn package

sns.boxplot(data = df1, x ='member_casual', y = 'ride_length', order = ['member', 'casual'])
```

```python
# Changing the datatype 
df1 = df1.astype({'ride_id':'string', 'rideable_type':'category', 'member_casual':'category'})
```

```python
df1.info()
```

```python
df1.shape
```

```python
df1['ride_id'] = df1['ride_id'].str.strip()
```

```python
df1[df1['ride_id'].duplicated()].any()
```

```python
df1['year'] = df1['started_at'].dt.year

cats1 = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df1['month'] = df1['started_at'].dt.month_name()
df1['month'] = df1['month'].astype(CategoricalDtype(categories=cats1, ordered=False))

cats2 = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
df1['day_of_week'] = df1['started_at'].dt.day_name()
df1['day_of_week'] = df1['day_of_week'].astype(CategoricalDtype(categories=cats2, ordered=False))

df1['hour'] = df1['started_at'].dt.hour

df1 = df1.astype({'year':'int16', 'hour':'int8'})

# Creating new columns "year", "month", "day_of_week", "hour" and Converting datatypes.

df1.head()
```

```python
# Line Graph of Month vs Ride Length

sns.lineplot(x='month', y='ride_length', data =  df1).set(title='Month vs Ride Length')
plt.show()

# You can see that summer is where customers prefer more bike rides compared to all seasons 
# Pricesly in the months of March,April,May,July
```

```python
# Saturday was the highest day Annual members and Casual riders rode bikes

df1['day_of_week'].value_counts()
```

```python
pd.pivot_table(df1,
              index = 'member_casual',
              values = 'ride_id',
              aggfunc = ['count'],
              margins = True,
              margins_name = 'Total Count')
```

```python
# You can see that we have more annual members compared to casual members

sns.countplot(df1['member_casual'])
plt.show()
```

```python
sns.countplot(x='day_of_week',data = df1, hue='member_casual').set(title='Day of the Week vs Member Count')
plt.show()
```

```python
# Visualizing days of the weeks vs types of rides

sns.countplot(x='day_of_week',data = df1[df1['member_casual'] == 'casual'], hue='rideable_type').set(title='Day of the Week vs Casual Rider Count')
plt.show()
```

#### Summary/Recommendations

1). Discounted memberships on weekends as you can see from the visualization customers prefer more rides on weekends. By this we can convert those casual members to annual members

2). You can clearly see that in the summer holidays customers rides increases alot compared to other seasons. Start giving discount memberships during holiday's which can again help boost our memberships

3). Reducing rent of the electric bikes for annual members can also be a good way to convert our casual customers and can also attract new customers

4). Fixed bike travel range or duration a non-member can do. This would help in new customers.

5). Every bike dock distance between one another should be within 1-2 mile radius. So, that customers can park and have no hard time parking.
